```python
#Regroupement des colonnes
def regrouper_Q(df, prefixe_question, nom_colonne_finale):
    """
    Regrouper les colonnes d'une question à choix multiples
    en une seule colonne contenant une liste, puis supprime les colonnes originales.
    
    Args:
        df : DataFrame
        prefixe_question : str, le préfixe commun (ex: "Q7")
        nom_colonne_finale : str, nom de la nouvelle colonne à créer (ex: "Languages")
        
    Returns:
        df mis à jour avec la nouvelle colonne et sans les colonnes originales
    """
  
    colonnes = [col for col in df.columns if col.startswith(prefixe_question + "_")]
    colonnes_parts = [col for col in colonnes if "Part" or "OTHER" in col]
    df[nom_colonne_finale] = df[colonnes_parts].apply(lambda x: [i for i in x if pd.notna(i)] if any(pd.notna(x)) else np.nan,axis=1)
    df.drop(columns=colonnes, inplace=True)
    
    return df
```

```python
df = regrouper_Q(df, "Q7", "Q7")
df = regrouper_Q(df, "Q9", "Q9")
df = regrouper_Q(df, "Q10", "Q10")
df = regrouper_Q(df, "Q12", "Q12")
df = regrouper_Q(df, "Q14", "Q14")
df = regrouper_Q(df, "Q16", "Q16")
df = regrouper_Q(df, "Q17", "Q17")
df = regrouper_Q(df, "Q18", "Q18")
df = regrouper_Q(df, "Q19", "Q19")
df = regrouper_Q(df, "Q23", "Q23")
df = regrouper_Q(df, "Q26_A", "Q26_A")
df = regrouper_Q(df, "Q27_A", "Q27_A")
df = regrouper_Q(df, "Q28_A", "Q28_A")
df = regrouper_Q(df, "Q29_A", "Q29_A")
df = regrouper_Q(df, "Q31_A", "Q31_A")
df = regrouper_Q(df, "Q33_A", "Q33_A")
df = regrouper_Q(df, "Q34_A", "Q34_A")
df = regrouper_Q(df, "Q35_A", "Q35_A")
df = regrouper_Q(df, "Q36", "Q36")
df = regrouper_Q(df, "Q37", "Q37")
df = regrouper_Q(df, "Q39", "Q39")
df = regrouper_Q(df, "Q26_B", "Q26_B")
df = regrouper_Q(df, "Q27_B", "Q27_B")
df = regrouper_Q(df, "Q28_B", "Q28_B")
df = regrouper_Q(df, "Q29_B", "Q29_B")
df = regrouper_Q(df, "Q31_B", "Q31_B")
df = regrouper_Q(df, "Q33_B", "Q33_B")
df = regrouper_Q(df, "Q34_B", "Q34_B")
df = regrouper_Q(df, "Q35_B", "Q35_B")

#Les colonnes ont été regroupé dans le désordre, donc cette ligne sert à les remettre dans l'ordre croissant des questions
colonnes_ordonnees = sorted(df.columns, key=lambda x: (x.startswith("Q"), int(x[1:].split("_")[0]) if x.startswith("Q") and x[1:].split("_")[0].isdigit() else float('inf')))
df = df[colonnes_ordonnees]
df.head()
```

```python
#Lors du regroupement, les lignes "vides" se sont complétés avec des "[]", donc j'ai remplacé ces crochets par des NaN pour que ce soit plus visible
df.replace("", np.nan, inplace=True)

#Suppression des colonnes qui ne me semble pas pertinentes par rapport à notre problématique, il faut qu'on se mette d'accord tous ensemble sur les questions à garder
df.drop(columns=["Time from Start to Finish (seconds)", 'Q8', "Q12", "Q13","Q17", "Q18", "Q20", "Q21", "Q24", "Q25", "Q26_B", "Q27_B", "Q28_B", "Q29_B", "Q31_B", "Q33_B", "Q34_B", "Q35_B"], inplace=True)
```
